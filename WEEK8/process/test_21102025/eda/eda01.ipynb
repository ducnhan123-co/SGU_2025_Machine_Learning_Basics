# Phân tích khám phá dữ liệu


```python
# Load libraries
import os, sys
from IPython import display
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import joblib
```

```python
df_train = pd.read_csv("../../../data/train.csv")
df_test = pd.read_csv("../../../data/test.csv")    
```

```python
# Fare range & distribution
fare = df_train['Fare'].dropna().astype(float)

print("Min:", fare.min())
print("Max:", fare.max())
print("Mean:", fare.mean())
print("Median:", fare.median())
print("Std:", fare.std())
print("\nQuantiles:")
print(fare.quantile([0, 0.01, 0.05, 0.25, 0.5, 0.75, 0.95, 0.99, 1.0]))
print("\nDescribe:")
display.display(fare.describe())

# Plots
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
sns.histplot(fare, bins=50, kde=True)
plt.title('Histogram Fare')

plt.subplot(1,2,2)
sns.boxplot(x=fare)
plt.title('Boxplot Fare')

plt.tight_layout()
plt.show()
#
```
