- String-Categorical: 2  
- Numeric-Boolean: 6

```python
import os
from IPython.display import SVG, display

plot_folder = "AutoViz_Plots/Response"

# Gom nhóm tên file theo loại biểu đồ
univariate_keywords = ["hist", "box", "bar"]   # đơn biến
multivariate_keywords = ["scatter", "corr", "pair"]  # đa biến


```

## Biểu đồ đơn biến

```python
print("📊 Biểu đồ đơn biến (Univariate Analysis):")
for file in sorted(os.listdir(plot_folder)):
    if file.endswith(".svg") and any(key in file.lower() for key in univariate_keywords):
        display(SVG(filename=os.path.join(plot_folder, file)))
```

📊 Giải thích các biểu đồ đơn biến (Univariate Analysis)
1. Histogram (Biểu đồ tần suất)

Mục đích: mô tả phân bố của một biến liên tục hoặc rời rạc.

Cách đọc:

Trục X: giá trị của biến.

Trục Y: số lần xuất hiện (tần suất).

Ý nghĩa:

Cho biết biến có phân phối chuẩn hay lệch trái/phải.

Phát hiện giá trị bất thường (outlier).

Ví dụ: thu nhập khách hàng tập trung nhiều ở mức trung bình, ít ở mức rất cao.

2. Boxplot (Biểu đồ hộp)

Mục đích: mô tả độ phân tán và phát hiện ngoại lệ.

Cách đọc:

Hộp: chứa 50% dữ liệu (từ Q1 đến Q3).

Đường giữa hộp: median (trung vị).

Râu (whisker): khoảng dữ liệu trong giới hạn 1.5 * IQR.

Điểm ngoài râu: ngoại lệ (outliers).

Ý nghĩa:

Giúp so sánh mức độ phân tán giữa các biến.

Xác định biến có nhiều ngoại lệ hay không.

3. Bar chart (Biểu đồ cột)

Mục đích: mô tả tần suất xuất hiện của các giá trị trong biến phân loại (categorical).

Cách đọc:

Trục X: các hạng mục (ví dụ: nghề nghiệp, tình trạng hôn nhân).

Trục Y: số lượng hoặc tỷ lệ.

Ý nghĩa:

Cho biết hạng mục nào phổ biến nhất.

So sánh trực quan tỷ lệ giữa các nhóm.

Ví dụ: khách hàng kết hôn chiếm nhiều hơn so với khách hàng độc thân.

📌 Tóm tắt

Histogram → Phân phối dữ liệu (chuẩn hay lệch).

Boxplot → Độ phân tán + ngoại lệ.

Bar chart → Tần suất của biến phân loại.
